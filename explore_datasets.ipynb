# Exploring Datasets for Mood-Based Recommendation System

This notebook loads and displays the datasets used in the mood-based recommendation system: songs, movies, and books.

```python
import pandas as pd
import os

# Set the data directory
data_dir = 'data'

# List of datasets
datasets = ['songs.csv', 'movie.csv', 'books.csv']

print("Available datasets:")
for dataset in datasets:
    print(f"- {dataset}")
```

## 1. Songs Dataset

```python
# Load and display songs dataset
songs_df = pd.read_csv(os.path.join(data_dir, 'songs.csv'))
print("Songs Dataset Shape:", songs_df.shape)
print("\nFirst 5 rows:")
songs_df.head()
```

```python
# Dataset info
songs_df.info()
```

## 2. Movies Dataset

```python
# Load and display movies dataset
movies_df = pd.read_csv(os.path.join(data_dir, 'movie.csv'))
print("Movies Dataset Shape:", movies_df.shape)
print("\nFirst 5 rows:")
movies_df.head()
```

```python
# Dataset info
movies_df.info()
```

## 3. Books Dataset

```python
# Load and display books dataset
books_df = pd.read_csv(os.path.join(data_dir, 'books.csv'), low_memory=False)
print("Books Dataset Shape:", books_df.shape)
print("\nFirst 5 rows:")
books_df.head()
```

```python
# Dataset info
books_df.info()
```
